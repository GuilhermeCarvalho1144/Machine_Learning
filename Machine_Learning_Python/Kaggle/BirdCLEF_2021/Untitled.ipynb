```python
#Import Libs
import librosa
import librosa.display
import numpy as np
import matplotlib.pyplot as plt
import IPython.display as ipd
import os
```

```python
#Load and plot an audio sample
audio_sample, sr = librosa.load('/media/guilherme/Data/Database/Kaggle/BirdCLEF/train_short_audio/acafly/XC109605.ogg')
D = librosa.amplitude_to_db(np.abs(librosa.stft(audio_sample)), ref=np.max)
fig, ax = plt.subplots(nrows=2)
librosa.display.waveplot(audio_sample, sr=sr, ax=ax[0])
librosa.display.specshow(D, y_axis='linear', x_axis='time', sr=sr, ax=ax[1])
```
