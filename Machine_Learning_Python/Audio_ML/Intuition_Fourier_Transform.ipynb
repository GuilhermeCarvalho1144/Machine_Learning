```python
import librosa
import librosa.display
import numpy as np
import matplotlib.pyplot as plt
from scipy import fft
from IPython.display import Audio
```

```python
audio_sample, sr_sample = librosa.load('audios/10 - Fourier Transform_ The Intuition_audio_piano_c.wav')
Audio(audio_sample, rate=sr_sample)
```

```python
librosa.display.waveplot(audio_sample)
```

```python
ft = fft.fft(audio_sample)
mag = np.absolute(ft)
freq = np.linspace(0, sr_sample, len(mag))
plt.plot(freq[:5000], mag[:5000])
```

```python
#deep look into the signal
plt.plot(audio_sample[10000:10400])
```

```python
#signal vs sine
f = 523
phase = 0.55
# t = librosa.samples_to_time(range(len(audio_sample)), sr_sample)
samples = range(len(audio_sample))
t = librosa.samples_to_time(samples, sr=sr_sample)

plt.figure(figsize=(10,8))

sin = .1*np.sin(2*np.pi*(f*t-phase))
plt.plot(t[10000:10400], audio_sample[10000:10400])
plt.plot(t[10000:10400], sin[10000:10400], color='r')

plt.fill_between(t[10000:10400], sin[10000:10400]*audio_sample[10000:10400], color='y')
```
